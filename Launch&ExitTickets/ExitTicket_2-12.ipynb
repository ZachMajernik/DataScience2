```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from IPython.display import clear_output
import random
%config NotebookApp.iopub_msg_rate_limit=10000
```

```python
df_1 = pd.read_csv('spam.csv')
df_1.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis=1, inplace=True)
df_1['v2'] = df_1['v2'].astype(str).str.lower()
df_1.info()
```

```python
random.seed(7)

#training set will be 70% of original df
numForTraining = round(0.7*len(df_1),0)
trainIDX = []
while len(trainIDX) < numForTraining:
    num = random.randint(0, len(df_1)-1)
    if not num in trainIDX:
        trainIDX.append(num)

testDF = df_1.drop(trainIDX).reset_index(drop=True)
trainDF = df_1.iloc[trainIDX].reset_index(drop=True)
```

```python
def findWords(hamOrSpam, column, df, num):
    locDF = df.loc[df.loc[:,'v1'] == hamOrSpam]

    allWords = []
    for scentance in locDF[column]:
        words = scentance.split()
        for word in words:
            allWords.append(word)

    dictionary = {'category':[], 'topWord':[], 'count':[], 'proportionOfAllWords':[]}
    word_counts = pd.DataFrame(dictionary)

    for word in allWords:
        if word in word_counts['topWord'].tolist():
            index = word_counts[word_counts['topWord'] == word].index[0]
            word_counts.loc[index, 'count'] += 1
        else:
            new_word = pd.DataFrame({'category':[hamOrSpam], 'topWord':[word], 'count':[1]})
            word_counts = pd.concat([word_counts, new_word], ignore_index=True)
    
    for j in range(len(word_counts)):
        word_counts.loc[j, 'proportionOfAllWords'] = word_counts.loc[j, 'count']/len(allWords)
         
    word_counts = word_counts.sort_values(by='count', ascending=False).reset_index(drop=True)
    word_counts = word_counts.drop('count', axis=1)  
    return word_counts.iloc[0:num]
```

```python
findWords('ham', 'v2', trainDF, 100)
```
